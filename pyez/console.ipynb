# Console connection using PyEZ


## Now PyEZ will support Console (Telnet/Serial) connection apart from SSH

1. Telnet Connection
   * Over managment ip
   * Over Console Server ip/port
2. Serial Connection
3. Examples includes
   * Facts Gathering
   * cli calls
   * rpc calls
   * Load configuration over Serial connection
   * Install image over Serial connection

### To get this feature you need to install PyEZ from github (Will be released in PyPI in coming days)
### using below command

```
pip install -U git+https://github.com/Juniper/py-junos-eznc.git
```

## Telnet over management IP

````python
from jnpr.junos import Device
from lxml import etree

with Device(host='10.xxx.1.xxx', user='xxx', password='xxxx', mode='telnet', port='23', gather_facts=True) as dev:
    print dev.facts
    print dev.cli("show version", format='text', warning=False)
    op = dev.rpc.get_interface_information(interface_name='fxp0')
    print etree.tounicode(op)
````

### Load config over Serial connection

#### Can use any PyEZ Config Class functionality

````python
from jnpr.junos import Device
from jnpr.junos.utils.config import Config
from lxml import etree
from jnpr.junos.utils.config import Config

import logging
logging.basicConfig(level=logging.INFO)

with Device(user='root', password='xxxxx', mode='serial', port='/dev/tty.usbserial', gather_facts=True) as dev:
    print dev.facts
    print dev.cli("show version", format='text', warning=False)
    op = dev.rpc.get_interface_information(interface_name='fxp0')
    print etree.tounicode(op)
    cu = Config(dev)
    cu.load(path='/var/tmp/chocmx.conf', format='text', overwrite=True)
    cu.pdiff()
    print "commit result: ", cu.commit()
````
